```python
# importing libraries
import pandas as pd
import numpy as np
import csv
import math
```

```python
csv_loc = 'AB_NYC_2019.csv'
```

```python
# read in dataset
df = pd.read_csv(csv_loc, header = 0, index_col = None, usecols=["neighbourhood_group", "latitude", "longitude", "room_type", "minimum_nights", "price", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count", "availability_365"])
df = df.dropna()
```

```python
def remove_outliers(df, q=0.05):
    upper = df.quantile(1-q)
    lower = df.quantile(q)
    mask = (df < upper) & (df > lower)
    return mask
```

```python
mask = remove_outliers(df['price'], 0.05)
df_no_outlier = df[mask]
sampled_df = df_no_outlier.sample(frac=0.5, replace=False, random_state=1)
```

```python
print(sampled_df)
```

```python
sampled_df.to_csv("AB_NYC_2019_cleaned.csv", header=["neighbourhood_group", "latitude", "longitude", "room_type", "price", "minimum_nights", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count", "availability_365"])
```
